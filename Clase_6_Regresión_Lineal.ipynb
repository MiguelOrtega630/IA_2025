```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
data=pd.read_excel("/content/Salary.xlsx")
data.head()
```

```python
df=pd.DataFrame(data) #Crear Dataframe
df.head()
```

```python
X=df[['YearsExperience']]
Y=df['Salary']

plt.figure(figsize=(16,8))
plt.scatter(X,Y, color= "pink", label="Datos Reales")
plt.title("Salario vs AÃ±os de experiencia")
plt.legend()
plt.grid(True)
plt.show()
```

```python
data.columns
X=data['YearsExperience']
Y=data['Salary']
model=LinearRegression()
model.fit(X.values.reshape(-1,1),Y)
y_pred=model.predict(X.values.reshape(-1,1))
r2=r2_score(Y,y_pred)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,Y, color= "pink", label="Datos Reales")
plt.plot(X,y_pred, color="blue",linewidth=2, label="RegresiÃ³n Lineal")
plt.ylabel("Salario")
plt.xlabel("AÃ±os de experiencia")
plt.title("Salario vs AÃ±os de experiencia")
plt.legend()
plt.grid(True)
plt.show()

print(f"Ecuacion de la recta y= {model.coef_[0]:.2f}x+ {model.intercept_:.2f}")
print(f"Coeficiente de DeterminaciÃ³n (R2):{r2:2f}")
```
