```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min

%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D

plt.rcParams["figure.figsize"] = (16,9)
plt.style.use("ggplot")
```

```python
dataframe = pd.read_csv("analisis.csv")
dataframe.head()
```

```python
dataframe.describe()
dataframe.groupby("categoria").size()
```

```python
dataframe.drop(columns=["categoria"]).hist()
plt.show()
```

```python
sb.pairplot(
    dataframe.dropna(),
    hue="categoria",
    height=4,
    vars=["ex", "co", "op"],
    kind="scatter",
    palette="tab10"
)
```

```python
x = np.array(dataframe[["ex", "co", "op"]])
y = np.array(dataframe["categoria"])
x.shape
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

colores_reales = ["blue","red","green","cyan","magenta","orange","black","pink","brown","purple"]
asignar_reales = [colores_reales[row] for row in y]

ax.scatter(x[:,0], x[:,1], x[:,2], c=asignar_reales, s=60)
ax.set_xlabel("Extraversion (ex)")
ax.set_ylabel("Cooperativeness (co)")
ax.set_zlabel("Openness (op)")
plt.show()
```

```python
Nc = range(1,20)
kmeans = [KMeans(n_clusters=i).fit(x) for i in Nc]
model = kmeans[4]

labels = model.predict(x)
c = model.cluster_centers_
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

colores_clusters = ["yellow","navy","lime","brown","gray"]
asignar_clusters = [colores_clusters[row] for row in labels]

ax.scatter(x[:,0], x[:,1], x[:,2], c=asignar_clusters, s=60)
ax.scatter(c[:,0], c[:,1], c[:,2], marker="*", c="red", s=1000)

ax.set_xlabel("Extraversion (ex)")
ax.set_ylabel("Cooperativeness (co)")
ax.set_zlabel("Openness (op)")
ax.set_title("Clusters obtenidos por K-Means (ex, co, op)")
plt.show()
```

```python
score = [kmeans[i].score(x) for i in range(len(kmeans))]

plt.plot(Nc, score)
plt.xlabel("Cantidad de Clusters")
plt.ylabel("Puntaje")
plt.title("Elbow Curve")
plt.show()
```

```python
copy = pd.DataFrame()
copy["usuario"] = dataframe["usuario"]
copy["categoria"] = dataframe["categoria"]
copy["label"] = labels

cantidadGrupo = pd.DataFrame()
cantidadGrupo["Cluster"] = list(range(5))
cantidadGrupo["Cantidad"] = copy.groupby("label").size()
cantidadGrupo
```

```python
group_referrer_index = copy["label"] == 0
group_referrals = copy[group_referrer_index]

diversidadGrupo = pd.DataFrame()
diversidadGrupo["categoria"] = sorted(copy["categoria"].unique())
diversidadGrupo["cantidad"] = group_referrals.groupby("categoria").size()
diversidadGrupo
```
